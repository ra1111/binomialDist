```python
# Let's plot a binomial distribution for given values of 'n' and 'p' in Python

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Set parameters
n = 50
p = 0.3

# Generate binomial data
x = np.arange(0, n+1)
y = binom.pmf(x, n, p)

# Plot binomial distribution
plt.figure(figsize=(10, 6))
plt.plot(x, y, 'o-')
plt.title('Binomial Distribution')
plt.xlabel('x')
plt.ylabel('Probability')
plt.grid(True)
plt.show()
```
